```python
import numpy as np
import pandas as pd
import random
from random import randint
from faker import Faker
import csv
```

```python
fake = Faker()
```

```python
def GenerateFakeData(n):
    fields = ['E-mail','Phone Number','Home Address','Persons Name','Year of birh','No of Kids','House','Annual Income','Speeding Tickets']
    PersonName=getPersonsName(n)
    email=getemail(n,PersonName)
    phoneNumber=getPhoneNumber(n)
    HomeAddress=getHomeAddress(n)
    Yearofbirth=GetyearofBirth(n)
    NoOfKids=getNofKids(n,Yearofbirth)
    housetype=getHouseType(n)
    income=getAnnualincome(n,NoOfKids,Yearofbirth,housetype)
    SpeedingTickets=getSpeedingTickets(n,Yearofbirth)
    
    data=[]
    for i in range(0,n):
        row=[]
        row.append(email[i])
        row.append(phoneNumber[i])
        row.append(HomeAddress[i])
        row.append(PersonName[i])
        row.append(Yearofbirth[i])
        row.append(NoOfKids[i])
        row.append(housetype[i])
        row.append(income[i])
        row.append(SpeedingTickets[i])
        data.append(row)
    
    filename = "FinalData.csv"
    with open(filename, 'w') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(fields)
        csvwriter.writerows(data)
   
```

```python
def getemail(n,name):
    rowlist=[]
    domains = [ "hotmail.com", "gmail.com", "aol.com", "mail.com" , "mail.kz", "yahoo.com"]
    ch=[0,1]
    for i in range(0,n):
        mail=random.choice(domains)
        j=random.choice(ch)
        res=name[i].split(" ")[j]+str(getPhoneseq(randint(1,5)))+'@'+mail
        rowlist.append(res)
    return rowlist
def getPhoneNumber(n):
    rowlist=[]
    for i in range(0,n+1):
        if (i <= n//2):
            number='+91 '+'9'+ str(getPhoneseq(9))
            rowlist.append(number)
        else:
            number='+91 '+'8'+ str(getPhoneseq(9))
            rowlist.append(number)
    return rowlist
    
def getPhoneseq(n):
    range_start = 10**(n-1)
    range_end = (10**n)-1
    return randint(range_start, range_end)

def getHomeAddress(n):
    rowlist=[]
    for i in range(1,n+1):
        rowlist.append(fake.address())
    return rowlist
def getPersonsName(n):
    rowlist=[]
    for i in range(0,n):
        first_name = fake.first_name()
        last_name = fake.last_name()
        rowlist.append(first_name+" "+ last_name)
    return rowlist
def GetyearofBirth(n):
    rowlist=[]
    for i in range(0,n):
        dob = ((fake.date_of_birth()).year)
        if dob >=2000:
            dob=dob-25
        if dob <=1940:
            dob=dob+25
        rowlist.append(dob)
    return rowlist

def getNofKids(n,dob):
    rowlist=[]
    for i in range(0,n):
        if dob[i]>1990:
            kids=randint(0,2)
            rowlist.append(kids)
        else:
            kids=randint(0,4)
            rowlist.append(kids)
    return rowlist
def getHouseType(n):
    rowlist=[]
    items = ['Rent','Own']
    for i in range(0,n):
        rowlist.append(random.choice(items))
    return rowlist
def getAnnualincome(n,Kids,yearofBir,House):
    rowlist=[]
    for i in range(0,n):
        if(House[i]=="Own" or (yearofBir[i] >=1970 and yearofBir[i] <=1990)):
            rowlist.append(randint(20000,40000))
        elif(House[i]=="Rent" or (yearofBir[i] >=1990 and yearofBir[i] <=1999)):
            rowlist.append(randint(5000,20000))
        elif(Kids[i] >=3 and House[i]=="Rent"):
            rowlist.append(randint(15000,25000))
        elif(House[i]=="Rent" and yearofBir[i] <=1970):
            rowlist.append(randint(1000,5000))
        else:
            rowlist.append(randint(3000,8000))
    return rowlist
           
def getSpeedingTickets(n,year):
    rowlist=[]
    for i in range(0,n):
        if(year[i]<=1950):
            rowlist.append(0)
        elif(year[i]>1950 and year[i]<=1970):
            rowlist.append(randint(0,2))
        elif(year[i]>1970 and year[i]<=1990):
            rowlist.append(randint(0,7))
        else:
            rowlist.append(randint(3,10))
    return rowlist
    
```
